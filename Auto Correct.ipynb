```python
def leve_util(string_a, string_b):
    '''
    Return:- The minimum number of modifications needed to
             equalize the two strings
    '''
    
    #Termination condition
    if((len(string_a)==0) or (string_a=='')):
        return len(string_b)
    
    if((len(string_b)==0) or (string_b=='')):
        return len(string_a)
    
    if(string_a[0] == string_b[0]):
        return leve_util(string_a[1:], string_b[1:])
    
    return 1 + min((leve_util(string_a[1:], string_b)), #Deletion of first character from A
                   (leve_util(string_a, string_b[1:])), #Insertion of first character in A
                   (leve_util(string_a[1:], string_b[1:]))) #Replacement of first character
```

```python
'''
02067400100
02067400119
Mob:- 8975044444
'''
def leve(string_a, string_b):
    '''
    Pre-process the strings before calculating the Levenshtein distance
    '''
    #Convert the string to lower case
    string_a, string_b = str.lower(string_a), str.lower(string_b)
    
    leve_distance = leve_util(string_a, string_b)
    
    return leve_distance

first_string, second_string = input().split()
leve(first_string, second_string)
```

```python
#Obtain the POS tags
POS_dict = defaultdict(set)

def create_POS_dict(word_corpus=word_corpus):
    global POS_dict
    #Remove punctuations
    for word in word_corpus:
        word = ''.join(str(char) for char in word if(char not in string.punctuation))
        tag = pos_tag([word])[0][1]
              
        POS_dict[tag].add(str.lower(word))
        
```

```python
def get_n_smallest(entry_list):
    smallest_num_1, smallest_num_2, smallest_num_3 = 3*[np.inf]
        
    for search_num in entry_list:
        if(search_num<=smallest_num_1):
            smallest_num_3 = smallest_num_2
            smallest_num_2 = smallest_num_1
            smallest_num_1 = search_num #5, 2, 1
            print("smallest_num_1:- %d" %(search_num))
        
            
    return smallest_num_1, smallest_num_2, smallest_num_3
    
```

```python
def get_n_nearest(word, corpus, n=3):
    
    similar_word_1, similar_word_2, similar_word_3 = 3*['']
    min_dist_1, min_dist_2, min_dist_3 = 3*[np.inf]
    
    for search_word in set(corpus):
        distance = leve(word, search_word)
        if(distance < min_dist_1):
            print(distance, search_word)
            min_dist_3, similar_word_3  = min_dist_2, similar_word_2
            min_dist_2, similar_word_2 = min_dist_1, similar_word_1
            min_dist_1, similar_word_1 = distance, search_word
        
            
    return similar_word_1, similar_word_2, similar_word_3
```

```python
def get_similar_word(test_word, k=3):
    '''
    Find the POS tag for the input word.
    Search for k(=3)-similar words in the corpus belonginf to the obtained POS tag.
    Input:- 
    test_word:- The word whose most similar word is to be found
    k:- k most similar words to the input word.
    
    Return:-
    k most similar words to the input word.
    '''
    #Remove punctuations
    test_word = ''.join(char for char in test_word if(char not in string.punctuation))
    
    #Get the pos tag
    tag = pos_tag([test_word])[0][1] #[(word, tag)]
    
    #Get the searching subset
    search_subset = POS_dict[tag]
    
    #Distance of test_word from each of the words in search_subset
    return get_n_nearest(test_word, search_subset, k)
```
